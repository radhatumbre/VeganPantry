```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import itertools
from requests.exceptions import ChunkedEncodingError

with open('proxies.txt', 'r') as f:
    proxies = [line.strip() for line in f if line.strip()]

if not proxies:
    print("No valid proxies found in proxies.txt")
else:
    print(f"Proxies loaded: {proxies}")

proxy_cycle = itertools.cycle(proxies)

df = pd.DataFrame()

```

```python
for proxy in proxies:
#     url = 'https://www.noracooks.com/category/meal-type/main-dish/page/1/'
    header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    url = f'https://www.noracooks.com/category/meal-type/main-dish/page/1/'
    proxy_dict = {'http': proxy, 'https': proxy}

    try:
        response = requests.get(url, headers=header, proxies=proxy_dict, timeout=5)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
    #         recipe_pg = extract(soup)
    #         temp_recipes = pd.concat([temp_recipes, recipe_pg], ignore_index=True)
            print(f"Successfully retrieved page  with proxy {proxy}")
        else:
            print(f"Failed to retrieve page with proxy {proxy}. Status code: {response.status_code}")
    except (requests.ConnectionError, requests.Timeout) as e:
        print(f"Proxy {proxy} failed for page : {e}")
    except ChunkedEncodingError as e:
        print(f"ChunkedEncodingError occurred for proxy {proxy} and page : {e}")
        continue
```

```python
title = []
# <h2 class="wprm-recipe-name wprm-block-text-none">Marry Me Chickpeas</h2>
for card in cards:
    proxy = next(proxy_cycle)
    link_tag = card.find('a')
    recipe_url = link_tag['href']
    header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    try:
        response = requests.get(recipe_url, headers=header)
        response.raise_for_status()  # Raise an exception for non-200 status codes
        temp_soup = BeautifulSoup(response.text, 'lxml')

        
        # Find the container for the nutrition labels
        recipe_name = temp_soup.find(class_='wprm-recipe-name')

        if recipe_name:
            title.append(recipe_name.text.strip())
        else:
            title.append('-')

   
    except requests.exceptions.RequestException as e:
        print(f"Failed to fetch the page: {e}")
        continue  # Move to the next card if there's an error

df['title'] = title
```
