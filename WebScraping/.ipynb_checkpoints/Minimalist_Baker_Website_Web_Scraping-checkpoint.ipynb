```python
recipes = pd.DataFrame()

# headers

title = []
prep_time = []
cook_time = []
total_time = []

course = []
cuisine = []
freezer_friendly = []
does_it_keep = []

recipe_ingredient = []
recipe_instructions = []
recipe_nutrition = []

for card in cards:
    link_tag = card.find('a')
    recipe_url = link_tag['href']
    try:
        response = requests.get(recipe_url, headers=headers)
        response.raise_for_status()  # Raise an exception for non-200 status codes
        temp_soup = BeautifulSoup(response.text, 'lxml')

        
        # Title
        h1_tag = temp_soup.find('h1')
        
        if h1_tag:
            title.append(h1_tag.text)
        else:
            title.append('-')
            
        
        
        # Time
        prep = temp_soup.find(class_='wprm-recipe-prep-time-container')
        cook = temp_soup.find(class_='wprm-recipe-cook-time-container')
        total = temp_soup.find(class_='wprm-recipe-total-time-container')
        
        # Extract and print prep time
        if prep:
            prep_time_span = prep.find(class_='wprm-recipe-details-minutes')
            prep_time_hours = prep.find(class_='wprm-recipe-details-hours')
            prep_time_minutes = prep_time_span.text.strip() if prep_time_span else ''
            prep_time_hours_text = prep_time_hours.text.strip() if prep_time_hours else ''
            
            prep_time_cleaned = f"{prep_time_hours_text} {prep_time_minutes}".strip()
            prep_time.append(prep_time_cleaned)
        else:
            prep_time.append('-')

        # Extract and print cook time
        if cook:
            cook_time_span = cook.find(class_='wprm-recipe-details-minutes')
            cook_time_hours = cook.find(class_='wprm-recipe-details-hours')
            cook_time_minutes = cook_time_span.text.strip() if cook_time_span else ''
            cook_time_hours_text = cook_time_hours.text.strip() if cook_time_hours else ''
            cook_time_cleaned = f"{cook_time_hours_text} {cook_time_minutes}".strip()
            cook_time.append(cook_time_cleaned)
        else:
            cook_time.append('-')

        # Extract and print total time
        if total:
            total_time_span = total.find(class_='wprm-recipe-details-minutes')
            total_time_hours = total.find(class_='wprm-recipe-details-hours')
            total_time_minutes = total_time_span.text.strip() if total_time_span else ''
            total_time_hours_text = total_time_hours.text.strip() if total_time_hours else ''
            total_time_cleaned = f"{total_time_hours_text} {total_time_minutes}".strip()
            total_time.append(total_time_cleaned)
        else:
            total_time.append('-')
            
            
            
        
        # Attributes
        course.append('-')
        cuisine.append('-')
        freezer_friendly.append('-')
        does_it_keep.append('-')
         # Find the div containing tags and values
        tags_div = temp_soup.find(class_='wprm-recipe-tags-container')

        # Extract and store values in lists
        if tags_div:
            tag_value_pairs = tags_div.find_all('div', class_='wprm-recipe-block-container')

            for pair in tag_value_pairs:
                label = pair.find(class_='wprm-recipe-tag-label').text.strip()
                value = pair.find(class_='wprm-block-text-normal').text.strip()

                if label == 'Course':
                    course[-1] = value
                elif label == 'Cuisine':
                    cuisine[-1] = value
                elif label == 'Freezer Friendly':
                    freezer_friendly[-1] = value
                elif label == 'Does it keep?':
                    does_it_keep[-1] = value

        
        
        
        # Ingredients
        # Find all <div> elements with class 'wprm-recipe-ingredient-group'
        ingredient_groups = temp_soup.find_all('div', class_='wprm-recipe-ingredient-group')
        if ingredient_groups:
            # Initialize an empty list to store ingredients
            all_ingredients_tags = []

            # Loop through each ingredient group
            for group in ingredient_groups:
                # Check if the group contains an <h4> tag with 'optional' in its text
                h4_tag = group.find('h4', class_='wprm-recipe-group-name')
                if h4_tag and 'optional' in h4_tag.text.lower():
                    continue
                else:
                    ingredients = group.find_all('li', class_='wprm-recipe-ingredient')
                    all_ingredients_tags.extend(ingredients)

            all_ingredients=[]
            for ingredient in all_ingredients_tags:
                all_ingredients.append(ingredient.text.strip())

            recipe_ingredient.append(all_ingredients)
        else:
            recipe_ingredient.append('-')
        
        
        # Instructions
        # Find the <div> containing the instructions
        instruction_group = temp_soup.find('div', class_='wprm-recipe-instruction-group')
        if instruction_group:
            # Initialize an empty list to store the instructions
            instructions = []

            # Find all <li> elements within the <ul> of instructions
            instruction_items = instruction_group.find_all('li', class_='wprm-recipe-instruction')

            # Loop through each instruction item
            for item in instruction_items:
                # Check if <span> exists within the item
                instruction_span = item.find('span')
                if instruction_span:
                    # Extract the text of the instruction
                    instruction_text = instruction_span.text.strip()
                    # Add the instruction to the list
                    instructions.append(instruction_text)

            recipe_instructions.append(instructions)
        else:
            recipe_instructions.append('-')
        
        # Nutrition
         # Find the container for the nutrition labels
        nutrition_container = temp_soup.find('div', class_='wprm-nutrition-label-container')
        if nutrition_container:
            # Initialize an empty dictionary to store the nutrition information
            nutrition_dict = {}

            # Find all the nutrition label text containers
            nutrition_labels = nutrition_container.find_all('span', class_='wprm-nutrition-label-text-nutrition-label')

            # Extract the nutrition labels and values
            for label in nutrition_labels:
                # Get the label text and value
                value_span = label.find_next('span', class_='wprm-nutrition-label-text-nutrition-value')
                value = value_span.text.strip()
                unit_span = value_span.find_next('span', class_='wprm-nutrition-label-text-nutrition-unit')
                unit = unit_span.text.strip()
                label_text = label.text.strip()

                # Add the label and value to the dictionary
                nutrition_dict[label_text] = f"{value} {unit}" 

            recipe_nutrition.append(nutrition_dict)
        else:
            recipe_nutrition.append('-')
        
    except requests.exceptions.RequestException as e:
        print(f"Failed to fetch the page: {e}")
        continue  # Move to the next card if there's an error


recipes['title'] = title
recipes['prep_time'] = prep_time
recipes['cook_time'] = cook_time
recipes['total_time'] = total_time

recipes['course'] = course
recipes['cuisine'] = cuisine
recipes['freezer_friendly'] = freezer_friendly
recipes['does_it_keep'] = does_it_keep
recipes['ingredients'] = recipe_ingredient
recipes['instructions'] = recipe_instructions
recipes['nutrition'] = recipe_nutrition
```

```python
ingredients = []

for card in cards:
    link_tag = card.find('a')
    recipe_url = link_tag['href']
    try:
        response = requests.get(recipe_url, headers=headers)
        response.raise_for_status()  # Raise an exception for non-200 status codes
        temp_soup = BeautifulSoup(response.text, 'lxml')

        
        # Find the container for the nutrition labels
        nutrition_container = temp_soup.find('div', class_='wprm-nutrition-label-container')

        # Initialize an empty dictionary to store the nutrition information
        nutrition_dict = {}

        # Find all the nutrition label text containers
        nutrition_labels = nutrition_container.find_all('span', class_='wprm-nutrition-label-text-nutrition-label')

        # Extract the nutrition labels and values
        for label in nutrition_labels:
            # Get the label text and value
            value_span = label.find_next('span', class_='wprm-nutrition-label-text-nutrition-value')
            value = value_span.text.strip()
            unit_span = value_span.find_next('span', class_='wprm-nutrition-label-text-nutrition-unit')
            unit = unit_span.text.strip()
            label_text = label.text.strip()
#             label_value = label.find_next('span', class_='wprm-nutrition-label-text-nutrition-value').text.strip()

            # Add the label and value to the dictionary
            nutrition_dict[label_text] = f"{value} {unit}" 

        # Print the dictionary
        print(nutrition_dict)

    

        print("\n----------------------------------\n")


    except requests.exceptions.RequestException as e:
        print(f"Failed to fetch the page: {e}")
        continue  # Move to the next card if there's an error



```
