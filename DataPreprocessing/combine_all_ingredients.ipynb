```python
inr = []
```

```python
with open('base_ingredients_lst.txt', 'r') as file:
    for line in file:
        inr.append(line.strip())
```

```python
df = pd.read_csv('nonvegan-alternatives.csv')
df
```

```python
def extract_altr(lst):
    temp = lst.split(', ')
    print(temp)
    for i in temp:
        if i not in inr:
            inr.append(i)
```

```python
df['VeganAlternative'].apply(extract_altr)
```

```python
def save_list_to_txt(filename, data_list):
    with open(filename, 'w') as f:
        for item in data_list:
            f.write(f"{item}\n")
```
