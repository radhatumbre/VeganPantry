```python
df = pd.read_csv('df_small.csv')
```

```python
import ast
df['ingredients'] = df['ingredients'].apply(ast.literal_eval)
```

```python
df
```

```python
def extract_base_ingr(ingredients_list):
    base_ingr = []
    for ingr in ingredients_list:
        ingr = ingr.lower()
        found = False
        for key in mapping_dict:
            if found == True:
                break
            if key.lower() in ingr:
                found = True
                base_ingr.append(mapping_dict[key].lower())
                
    # remove duplicates
    seen = set()
    base_ingr = [x for x in base_ingr if not (x in seen or seen.add(x))]
    return base_ingr
    
df['base_ingredients'] = df['ingredients'].apply(extract_base_ingr)
```

# Create a list of all the base ingredients in the DataBase

```python
base_ingredients_lst = []
for key in mapping_dict:
    base_ingredients_lst.append(mapping_dict[key].lower())

# remove duplicates
no_duplicate_base_ingr = set()
base_ingredients_lst = [x for x in base_ingredients_lst if not (x in no_duplicate_base_ingr or no_duplicate_base_ingr.add(x))]
```

```python
len(base_ingredients_lst)
```

```python
base_ingredients_lst.sort()
```

```python
def save_list_to_txt(filename, data_list):
    with open(filename, 'w') as f:
        for item in data_list:
            f.write(f"{item}\n")
```

```python
save_list_to_txt('base_ingredients_lst.txt', base_ingredients_lst)
```
