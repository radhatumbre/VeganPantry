```python
import ast
def safe_literal_eval(val):
    if val is None:
        return val
    try:
        return ast.literal_eval(val)
    except (ValueError, SyntaxError):
        return val
    
# Convert string representations of lists to actual lists
# df_small['ingredients'] = df_small['ingredients'].apply(ast.literal_eval)
df['ingredients'] = df['ingredients'].apply(safe_literal_eval)
```

```python
all_ingredients =[]
```

```python
base_ingredients = ['steak', 'chicken','beef']

```

```python
def extract_ingr(ingr):
    global all_ingredients  # Use global to modify the list outside the function
    if ingr is not None:
        for i in ingr:
            if 'for serving' in i:
                continue
            words = [word.strip() for word in i.replace(',', ' ').split()]
            for word in words:
                if word in base_ingredients:
                    all_ingredients.append(word)
                    break  # Exit the loop once a match is found
            else:
                all_ingredients.append(i)  # If no match found, add the entire ingredient
```

```python
def save_list_to_txt(filename, data_list):
    """
    Save a list of strings to a text file, one item per line.
    
    :param filename: The name of the file to save the list to.
    :param data_list: The list of strings to save.
    """
    with open(filename, 'w') as f:
        for item in data_list:
            f.write(f"{item}\n")
```
